```python
import nltk
import urllib.request
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
```

```python
response = urllib.request.urlopen('http://php.net/') 
html = response.read()
soup=BeautifulSoup(html,'html5lib')
text=soup.get_text(strip=True)
```

```python
tokens=[t for t in text.split()]

cleant=tokens[:]
sw=stopwords.words('english')
for i in tokens:
    if i in sw:
        cleant.remove(i)
        
freq=nltk.FreqDist(cleant)
freq
```

```python
# POS tag list:

# CC	coordinating conjunction
# CD	cardinal digit
# DT	determiner
# EX	existential there (like: "there is" ... think of it like "there exists")
# FW	foreign word
# IN	preposition/subordinating conjunction
# JJ	adjective	'big'
# JJR	adjective, comparative	'bigger'
# JJS	adjective, superlative	'biggest'
# LS	list marker	1)
# MD	modal	could, will
# NN	noun, singular 'desk'
# NNS	noun plural	'desks'
# NNP	proper noun, singular	'Harrison'
# NNPS	proper noun, plural	'Americans'
# PDT	predeterminer	'all the kids'
# POS	possessive ending	parent\'s
# PRP	personal pronoun	I, he, she
# PRP$	possessive pronoun	my, his, hers
# RB	adverb	very, silently,
# RBR	adverb, comparative	better
# RBS	adverb, superlative	best
# RP	particle	give up
# TO	to	go 'to' the store.
# UH	interjection	errrrrrrrm
# VB	verb, base form	take
# VBD	verb, past tense	took
# VBG	verb, gerund/present participle	taking
# VBN	verb, past participle	taken
# VBP	verb, sing. present, non-3d	take
# VBZ	verb, 3rd person sing. present	takes
# WDT	wh-determiner	which
# WP	wh-pronoun	who, what
# WP$	possessive wh-pronoun	whose
# WRB	wh-abverb	where, when
```

```python
from nltk.corpus import state_union #txt file
from nltk.tokenize import PunktSentenceTokenizer

train=state_union.raw("2005-GWBush.txt")
test=state_union.raw("2006-GWBush.txt")
test
```

```python
sentence_tokenizer=PunktSentenceTokenizer(train)
listofsentences=sentence_tokenizer.tokenize(test)
```

```python
def process_content():
    try:
        for i in listofsentences[:5]:
            indivwords=nltk.word_tokenize(i)
            tagged=nltk.pos_tag(indivwords)
            chunkg=r'''Chunk:{<RB.?>*<VB.?>*<NNP>+<NN>?}
                        }<VB.?|IN|DT|TO>+{'''
            chunkp=nltk.RegexpParser(chunkg)
            chunked=chunkp.parse(tagged)
            #print(chunked)
            for subtree in chunked.subtrees(filter=lambda t:t.label()=='Chunk'):
                print(subtree)
    except Exception as e:
        print(str(e))
process_content()
```

```python
def process_content():
    try:
        for i in listofsentences[:5]:
            indivwords=nltk.word_tokenize(i)
            tagged=nltk.pos_tag(indivwords)
            namedent=nltk.ne_chunk(tagged,binary=True) #classifies named entities (in brackets with 'NE' in front)
            #if binary=False, named entities will be catergorised
            print(namedent)
    except Exception as e:
        print(str(e))
process_content()
```
